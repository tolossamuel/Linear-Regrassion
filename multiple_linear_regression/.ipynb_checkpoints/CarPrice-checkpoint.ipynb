```python
import pandas as pd
import matplotlib.pyplot as plt
```

# extract data

```python
data_address = "datasets/CarPrice_Assignment.csv"
data_set = pd.read_csv(data_address)
```

```python
data_set.head()
```

```python
data_set.columns
```

# data preprocess

```python
fueltypes = {"gas" : 0, "diesel":1}
data_set["fueltype"] = [fueltypes[t] for t in data_set["fueltype"]]
aspirations = {"std":0, "turbo":1}
data_set["aspiration"] = [aspirations[a] for a in data_set["aspiration"]]
```

```python
data_set.head()
```

```python
Number = {'one': 1, 'two': 2, 'three': 3, 'four': 4, 'five': 5,
    'six': 6, 'seven': 7, 'eight': 8, 'nine': 9, 'ten': 10,
    'eleven': 11, 'twelve': 12, 'thirteen': 13, 'fourteen': 14, 'fifteen': 15,
    'sixteen': 16, 'seventeen': 17, 'eighteen': 18, 'nineteen': 19, 'twenty': 20
}
data_set["doornumber"] = [Number[d] for d in data_set["doornumber"]]
```

```python
data_set.head()
```

```python
driveWheel = {"rwd":2,"fwd":1,"4wd":3}
data_set["drivewheel"] = [driveWheel[w] for w in data_set["drivewheel"]]
data_set.head()
```

```python
engineLocations = {"front":1,"rear":0}
data_set["enginelocation"] = [engineLocations[e] for e in data_set["enginelocation"]]
data_set.head()
```

```python
data_set["cylindernumber"] = [Number[n] for n in data_set["cylindernumber"]]
data_set.head()
```

```python
fuelSystem = {"mpfi" : 0, "2bbl":1,"mfi":3, "1bbl":4, "spfi":5, "4bbl":6, "idi":7, "spdi": 8}
data_set["fuelsystem"] = [fuelSystem[f] for f in data_set["fuelsystem"]]
data_set.head()
```

```python
enginetypes = {"dohc":0, "ohcv":1, "ohc":3, "l":4, "rotor":5, "dohcv":6, "ohcf":7}
data_set["enginetype"] = [enginetypes[e] for e in data_set["enginetype"]]
data_set.head()
```

# data combination 

```python
# data_set["HR_P_NC"] = data_set["horsepower"]/data_set["cylindernumber"]
```

```python

independent = ["fueltype","aspiration","doornumber","drivewheel","drivewheel","wheelbase","carlength","carwidth","carheight",
               "curbweight","cylindernumber","enginesize","boreratio","stroke","compressionratio","horsepower","peakrpm","citympg",
               "highwaympg","fuelsystem","enginetype"]
X = data_set[independent]
y = data_set["price"]
```

```python
X.head()
```

<!--correlation between the price and x data  -->

# correlation between the y and x

```python
concatenated_df = pd.concat([X, y], axis=1)
corr_matrix = concatenated_df.corr()

corr_matrix["price"].sort_values(ascending = False)
```

# split the data for train and test

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y, test_size = 0.25,random_state = 45)
```

# linear regrassion model test

```python
from sklearn.linear_model import LinearRegression

LinearModel = LinearRegression()
LinearModel.fit(x_train,y_train)
```

# predict and test our linear regression model on test data

```python
predictData = LinearModel.predict(x_test)
plt.scatter(y_test,predictData)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error

MAE = mean_absolute_error(y_test,predictData)
r2 = r2_score(y_test,predictData)
r2
```

```python
for i in range(10):
    print(f"predict values {predictData[i].round().astype(int)} vs acctual values {y_test.iloc[i]}")
```

# ridge model test

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV
ridge_model = Ridge()
params = {"alpha":[1e-22,1e-17,1e-15,1e-6,1e-5,1e-3,1e-2,1,2,5,10,20]}
ridge_regressor = GridSearchCV(ridge_model,params,scoring = "neg_mean_squared_error", cv = 3)
```

```python
# fit the data to the model
ridge_regressor.fit(x_train,y_train)
```

# pridict with the model

```python
predictRidgeModel = ridge_regressor.predict(x_test)
plt.scatter(y_test,predictRidgeModel)
```

# test the model

```python
MeanAbsolutError = mean_absolute_error(y_test,predictRidgeModel)
r_squeare = r2_score(y_test,predictRidgeModel)
r_squeare
```

# lasso model 

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
lasso = Lasso()
params = {"alpha":[1e-50,1e-40,1e-25,1e-6,1e-5,1e-3,1e-2,1,2,5,10,20]}
lasso_regressor = GridSearchCV(lasso,params,scoring = "neg_mean_squared_error", cv = 3)
```

```python
# fit the data to the model
```

```python
lasso_regressor.fit(x_train,y_train)
```

# test the lasso model

```python
perdict_lasso = lasso_regressor.predict(x_test)

plt.scatter(y_test,perdict_lasso)
```

```python
lasso_mae = mean_absolute_error(y_test,perdict_lasso)
lasso_r2 = r2_score(y_test,perdict_lasso)
lasso_r2
```

# decision tree model

```python
from sklearn.tree import DecisionTreeRegressor
decisionModel = DecisionTreeRegressor()
decisionModel.fit(x_train,y_train)
```

```python
# predict the values
decisionPredict = decisionModel.predict(x_test)
plt.scatter(y_test,decisionPredict)
```

```python
decision_MAE = mean_absolute_error(y_test,decisionPredict)
decision_r2 = r2_score(y_test,decisionPredict)
decision_MAE
```

```python
for i in range(5):
    print(f"predict values {decisionPredict[i].round().astype(int)} vs acctueal values  {y_test.iloc[i]}")
```

# Random Forest Regressor model

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(x_train,y_train)
```

```python
# predict the data
forestPredict = regressor.predict(x_test)
```

```python
# test the model

mse = mean_squared_error(y_test, forestPredict)
print(f'Mean Squared Error: {mse}')
r2 = r2_score(y_test, forestPredict)
print(f'R-squared: {r2}')
```
